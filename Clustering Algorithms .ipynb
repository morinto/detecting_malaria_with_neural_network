```python
import numpy as np
from pyclustering.cluster.clarans import clarans
import pickle
```

```python
with open('all_img', 'rb') as f:
    imgs = pickle.load(f)
```

```python
imgs[0].shape
```

```python
img_list = []

for rows in imgs:
    r = rows.flatten()
    r = r.tolist()
    img_list.append(r)
```

```python
img_list[0]
```

```python
clarans_obj = clarans(img_list, 2, 4, 5)
clarans_obj.process()
```

```python
print(clarans_obj.get_clusters())
```
